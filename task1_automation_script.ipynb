```python
import os
import shutil
import csv

# Base directory setup
BASE_DIR = './file_demo'
TEXT_FILE = os.path.join(BASE_DIR, 'sample.txt')
CSV_FILE = os.path.join(BASE_DIR, 'data.csv')
ARCHIVE_FILE = os.path.join(BASE_DIR, 'archived_sample.txt')


def create_directory(path):
    if not os.path.isdir(path):
        os.makedirs(path)
        print(f"Created directory: {path}")


def create_text_file(filepath):
    with open(filepath, 'w') as file:
        file.write("Hello, this is a sample text file for automation demo.")


def create_csv_file(filepath):
    with open(filepath, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerows([
            ['ID', 'Name', 'Status'],
            [1, 'Task_A', 'Completed'],
            [2, 'Task_B', 'Pending']
        ])


def setup_environment():
    try:
        create_directory(BASE_DIR)
        create_text_file(TEXT_FILE)
        create_csv_file(CSV_FILE)
        print("Sample files created successfully.")
    except Exception as err:
        print(f"Error during setup: {err}")


def read_text_file(filepath):
    print("\nReading File")
    with open(filepath, 'r') as file:
        data = file.read()
        print(f"Content read: {data}")


def move_and_rename_file(src, dest):
    print("\nAutomating File Operations")
    if os.path.isfile(src):
        shutil.move(src, dest)
        print(f"Moved and Renamed '{src}' to '{dest}'")


def read_csv_file(filepath):
    print("\nReading CSV Data")
    with open(filepath, 'r') as file:
        reader = csv.DictReader(file)
        for record in reader:
            print(f"Processing: {record['Name']} - {record['Status']}")


def run_automation():
    try:
        read_text_file(TEXT_FILE)
        move_and_rename_file(TEXT_FILE, ARCHIVE_FILE)
        read_csv_file(CSV_FILE)

    except FileNotFoundError:
        print("Error: The requested file was not found.")
    except PermissionError:
        print("Error: You do not have permission to access these files.")
    except Exception as err:
        print(f"An unexpected error occurred: {err}")
    finally:
        print("\nFile operation sequence completed.")


def show_directory_contents(path):
    if os.path.exists(path):
        print(f"Current files in {path}:")
        print(os.listdir(path))


# Execute
setup_environment()
run_automation()
show_directory_contents(BASE_DIR)
```
